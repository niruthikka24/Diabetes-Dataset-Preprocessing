```python
df = pd.read_csv('diabetes.csv')
print(df.head())
print(df.shape)
print(df.isna().sum())
```

```python
df = df.drop_duplicates(subset=['Pregncies', 'Glucose', 'BloodPressure','SkinThickness','Insulin','BMI','DiabetesPedigreeFunction','Age','Outcome'])
df.shape
```

```python
df[['Glucose','BloodPressure','SkinThickness','Insulin','BMI']] = df[['Glucose','BloodPressure','SkinThickness','Insulin','BMI']].replace(0,np.NaN)
df.isnull().sum()
```

```python
df.describe()
```

```python
dfmed = df.copy(deep = True)
dfmed.head()
```

```python
dfmed.isna().sum()
```

```python
dfmed[dfmed < 0] = np.NaN #making negative values NaN
```

```python
dfmed.isna().sum()
```

```python
for col in dfmed.columns:
  dfmed[col] = dfmed[col].fillna(dfmed[col].median())
```

```python
dfmed.isna().sum()
```

```python
dfmed.describe()
```

```python
corr = dfmed.corr()
corr.style.background_gradient(cmap='coolwarm')
```

```python
dfmedout = dfmed.copy(deep=True)
dfmedout.isna().sum()
```

```python
for col in dfmedout.columns:
  Q1 = dfmedout[col].quantile(0.25)
  Q3 = dfmedout[col].quantile(0.75)
  IQR = Q3 - Q1
  whisker_width = 1.5
  lower_whisker = Q1 -(whisker_width*IQR)
  upper_whisker = Q3 + (whisker_width*IQR)
  med = dfmedout[col].median()
  dfmedout[col]=np.where(dfmedout[col]>upper_whisker,med,np.where(dfmedout[col]<lower_whisker,med,dfmedout[col]))
```

```python
dfmedout.describe()
```

```python
corr = round(dfmedout.corr(),2)
corr.style.background_gradient(cmap='coolwarm')
```

```python
dfclassmed = df.copy(deep = True)
dfclassmed[dfclassmed < 0] = np.NaN
dfclassmed.isna().sum()
```

```python
outcome0 = dfclassmed[dfclassmed['Outcome']==0]
outcome1 = dfclassmed[dfclassmed['Outcome']==1]
for col in dfclassmed.columns:
  med0 = outcome0[col].median()
  med1 = outcome1[col].median()
  missing = dfclassmed[col].isna()
  dfclassmed.loc[missing, col] = np.where(dfclassmed.loc[missing, 'Outcome'] == 0, med0, med1)
  # missing = dfclassmed[col].isna()
  # mapdict = dict({0:med0,1:med1})
  # missing.loc[missing,col] = dfclassmed.loc[missing,'Outcome'].map(mapdict)
  # dfclassmed[col] = dfclassmed[col].fillna()
```

```python
dfclassmed.isna().sum()
```

```python
dfclassmed.describe()
```

```python
outcome0 = dfclassmed[dfclassmed['Outcome']==0]
outcome1 = dfclassmed[dfclassmed['Outcome']==1]
for col in dfclassmed.columns:
  for target in [0,1]:
    dfsub = dfclassmed[dfclassmed['Outcome']==target]
    med = dfsub[col].median()
    Q1 = dfsub[col].quantile(0.25)
    Q3 = dfsub[col].quantile(0.75)
    IQR = Q3 - Q1
    whisker_width = 1.5
    lower_whisker = Q1 -(whisker_width*IQR)
    upper_whisker = Q3 + (whisker_width*IQR)
    dfclassmed.loc[(dfclassmed['Outcome'] == target) & (dfclassmed[col] < lower_whisker), col] = med
    dfclassmed.loc[(dfclassmed['Outcome'] == target) & (dfclassmed[col] > upper_whisker), col] = med
```

```python
dfnew = pd.read_csv('diabetes (1).csv')
dfnew.isna().sum()
```

```python
dfnew.drop_duplicates(subset=dfnew.columns.difference(['Patient_ID']),inplace=True)
dfnew.shape
```

```python
dfnew = dfnew.drop_duplicates(subset=['Pregncies', 'Glucose', 'BloodPressure','SkinThickness','Insulin','BMI','DiabetesPedigreeFunction','Age','Outcome'])
dfnew.shape
```

```python
dfnew[['Glucose','BloodPressure','SkinThickness','Insulin','BMI']] = dfnew[['Glucose','BloodPressure','SkinThickness','Insulin','BMI']].replace(0,np.NaN)
dfnew.isnull().sum()
```

```python
dfclassmedqr = dfnew.copy(deep=True)
dfclassmedqr[dfclassmedqr < 0] = np.NaN
dfclassmedqr.isna().sum()
```

```python
outcome0 = dfclassmedqr[dfclassmedqr['Outcome']==0]
outcome1 = dfclassmedqr[dfclassmedqr['Outcome']==1]
for col in dfclassmedqr.columns:
  median0 = outcome0[col].median()
  median1 = outcome1[col].median()
  missing = dfclassmedqr[col].isna()
  dfclassmedqr.loc[missing, col] = np.where(dfclassmedqr.loc[missing, 'Outcome'] == 0, median0, median1)
```

```python
for col in dfclassmedqr.columns:
  for target in [0,1]:
    dfsub = dfclassmedqr[dfclassmedqr['Outcome']==target]
    Q1 = dfsub[col].quantile(0.25)
    Q3 = dfsub[col].quantile(0.75)
    IQR = Q3 - Q1
    whisker_width = 1.5
    lower_whisker = Q1 -(whisker_width*IQR)
    upper_whisker = Q3 + (whisker_width*IQR)
    dfclassmedqr.loc[(dfclassmedqr['Outcome'] == target) & (dfclassmedqr[col] < lower_whisker), col] = lower_whisker
    dfclassmedqr.loc[(dfclassmedqr['Outcome'] == target) & (dfclassmedqr[col] > upper_whisker), col] = upper_whisker
```

```python
dflis = [dfmed,dfmedout,dfclassmed,dfclassmean,dfclassmedqr,dfpre3]
for df in dflis:
  x = df.drop(['Outcome','Patient_ID'],axis=1)
  y = df['Outcome']
  xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size=0.20, random_state=17)
  rf = RandomForestClassifier(n_estimators = 70)
  rf.fit(xtrain,ytrain)
  pred = rf.predict(xtest)
  acc = accuracy_score(ytest, pred)
  f1 = f1_score(ytest,pred)
  print("Accuracy", acc)
  print("F1", f1)
  print("=======================================================================")
```
