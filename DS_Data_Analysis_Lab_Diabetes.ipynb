```python
dfsummary = df.loc[:,['Insulin','BMI','Age']]
dfsummary.head()
```

```python
dfsummary.describe().T
```

```python
dfsummary.columns
```

```python
minlis = []
q1lis = []
q2lis = []
q3lis = []
maxlis = []
collis = []
for col in dfsummary.columns:
  collis.append(col)
  minlis.append(dfsummary[col].min())
  quartiles = percentile(dfsummary[col], [25, 50, 75])
  q1lis.append(quartiles[0])
  q2lis.append(quartiles[1])
  q3lis.append(quartiles[2])
  maxlis.append(dfsummary[col].max())
```

```python
dfsummary.describe()
```

```python
summary = pd.DataFrame(list(zip(collis,minlis,q1lis,q2lis,q3lis,maxlis)), columns =['Feature', 'Min','Lower Quartile','Median','Upper Quartile','Max'])
summary
```
